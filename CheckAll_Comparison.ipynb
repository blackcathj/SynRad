```python
import pickle
import pandas as pd
import os
import matplotlib.pyplot as plt
import matplotlib.colors
import numpy as np
import pickle
```

```python
base_dir1 = 'May2020Lattice/CharlieGeometry'
base_dir2 = 'May2020Lattice/UpdatedGeometryPumps'
```

```python
fid = open(f'{base_dir1}/flux_for_photon.pickle', 'rb')
hhistset0 = pickle.load(fid)
hhistset = pickle.load(fid)
hhistset2 = pickle.load(fid)


histsets1 = (hhistset0 , hhistset , hhistset2)
print(histsets1)
```

```python

fid = open(f'{base_dir2}/flux_for_photon.pickle', 'rb')
hhistset0 = pickle.load(fid)
hhistset = pickle.load(fid)
hhistset2 = pickle.load(fid)


histsets2 = (hhistset0 , hhistset , hhistset2)
print(histsets2)
```

```python

plt.figure(num=None, figsize=(8, 10), dpi=80, facecolor='w', edgecolor='k')

ax_p = plt.subplot(2,1,1)
ax_p.set_yscale('log')
ax_p.set_ylim((1e8,1e15))

ax_n = plt.subplot(2,1,2)
ax_n.set_yscale('log')
ax_n.set_ylim((1e8,1e15))

for histset in histsets1:

    ax_p.step(
        histset[3][0:-1],
        histset[2],
        label = histset[1]
    )
    ax_n.step(
        histset[5][0:-1],
        histset[4],
        label = histset[1]
    )

for histset in histsets2:

    ax_p.step(
        histset[3][0:-1],
        histset[2],
        label = histset[1]
    )
    ax_n.step(
        histset[5][0:-1],
        histset[4],
        label = histset[1]
    )

    
ax_p.legend()
ax_p.set_title("flux for photon | x>0")
ax_n.set_title("flux for photon | x<0")

ax_p.set_ylabel("Sum[Flux] photon/s/cm")
ax_n.set_ylabel("Sum[Flux] photon/s/cm")
ax_n.set_xlabel("z position @ photon crossing ref. facet [cm]")
```
